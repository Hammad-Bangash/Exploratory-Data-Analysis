Quantile-Quantile plot is used to determine whether two random continuous variables belong to same distribution or if we want to determine to which distribution a random variable belongs too.
So. for this purpose we calculate percentiles of the random variable and then plot it with the percentile values of the distribution. If the plot contains the points on the straight line then both belongs to the same class of distribution

```python
import numpy as np
import scipy.stats as stats
import pylab
import matplotlib.pyplot as plt

nor_dist=np.random.normal(loc=1.0,scale=4,size=5000)
stats.probplot(nor_dist,dist='norm',plot=pylab)
pylab.show()
print((
```

```python
nor_dist=np.random.normal(loc=1.0,scale=4,size=5000)
stats.probplot(nor_dist,dist='uniform',plot=pylab)
pylab.show()
```

```python
counts,bins=np.histogram(nor_dist,bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
plt.show()

```
