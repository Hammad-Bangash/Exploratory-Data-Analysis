```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(0.,5.,0.2)
print(a)

plt.plot(a,a**2,'rs',label="2")
plt.plot(a,a**2.2,'y^',label="2.2")
plt.plot(a,a**2.5,'b--',label="2.5")
plt.title("power")
plt.xlabel("numbers")
plt.ylabel("squares")
plt.legend()
plt.grid()
plt.show()
```

```python
x1=[1,2,3]
y1=[1,4,9]
y2=[2,4,6]
lines=plt.plot(x1,y1,x1,y2)
plt.setp(lines[0],color="r",linewidth=2)
plt.setp(lines[0],color="black",linewidth=3)
plt.grid()
plt.show()
```
