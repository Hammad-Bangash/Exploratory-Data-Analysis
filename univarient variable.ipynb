```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('iris_csv.csv')
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='class',height=4,).map(sns.distplot,'petallength').add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='class',height=4).map(sns.distplot,'petalwidth').add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='class',height=4).map(sns.distplot,'sepallength').add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='class',height=4).map(sns.distplot,'sepalwidth').add_legend()
plt.show()
```
