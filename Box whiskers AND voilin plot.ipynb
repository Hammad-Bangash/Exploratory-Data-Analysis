Box whiskers are used to plot data based on percentiles and quartile.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris=pd.read_csv('Iris_csv.csv')
sns.boxplot(x='class', y='petallength',data=iris)
plt.grid()
plt.show()
```

Here the lines are percentiles. The horizontal lines are called whiskers they are calculated by max and min value. But here in seaborn it is calculated as 1.5 x IQR. where IQR is the difference of 75th and 25th percentile. 
here in box the lowest line represents the 25th percent
ile,middle line represents the 50th percenitle and the upper line shows the 75th percentile. 
Through, these boxes we can make model and also estimate how percent of values it give will be correct and wrong.

Now, here is voilin plot which combines the histogram with the boxplot. where histogram shows the pdf and boxplot shows the percentile of datapoints. 
Here in voilin plot the white dot represents the 50th percentile.


```python
sns.violinplot(x='class',y='petallength',data=iris)
plt.grid()
plt.show()
```

```python
sns.FacetGrid(data=iris,hue='class',height=4).map(sns.distplot,'petallength').add_legend()
plt.show()
```
